```python
def load_data(file):
    df = pd.read_csv(os.path.join("data", file))
    return df

df = load_data("train.csv")
```

```python
import pandas as pd
from shapely.geometry import Point

df['Location'] = df.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

success_outcomes = ['Community resolution', 'Khat or Cannabis warning', 'Caution (simple or conditional)','Arrest', 'Penalty Notice for Disorder', 'Summons / charged by post',                     'Suspect arrested', 'Suspect summoned to court']

successful_searches = df[df['Outcome'].isin(success_outcomes) & df['Outcome linked to object of search'] == True]
unsuccessful_searches = df[(df['Outcome'].isin(success_outcomes) == False) &(df['Outcome'].isin(success_outcomes) & df['Outcome linked to object of search'] == False)]

counts = [len(successful_searches), len(unsuccessful_searches)]
labels = ['Successful search', 'Unsuccessful search']

plt.bar(labels, counts)

for i, count in enumerate(counts):
    plt.text(i, count+5000, str(round(count/sum(counts)*100,2)) + "%", ha='center')

plt.xlabel('Outcome')
plt.ylabel('Search Count')
plt.title('Successful vs Unsuccessful Searches')

plt.show()

```

## Outcome? como prediction


**Outcome Descriptions**

- Community resolution: A non-legal, non-punitive way of resolving minor crimes or incidents, where the offender accepts responsibility and agrees to take steps to make amends.


- A no further action disposal: The police have decided not to take any further action against the individual, and the case is closed.


- Arrest: The individual has been arrested and taken into custody for further questioning or charging.


- Summons / charged by post: The individual has been issued with a summons to attend court, or charged by post.


- Caution (simple or conditional): The individual has been given a formal warning by the police, which may be used as evidence in court if they re-offend.


- Penalty Notice for Disorder: The individual has been issued with a fine for a minor offence, such as disorderly behaviour.


- Khat or Cannabis warning: The individual has been given a verbal warning for possession of khat or cannabis.

```python
# convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')

# Set date as index and count number of searches per day
searches_by_date = df.set_index('Date').groupby(pd.Grouper(freq='D')).size()

# Create evenly distributed date range
date_range = pd.date_range(min(searches_by_date.index), max(searches_by_date.index), freq='D')

# Reindex searches_by_date with evenly distributed date range
searches_by_date = searches_by_date.reindex(date_range, fill_value=0)

# Create bar plot
plt.figure(figsize=(10,5))
plt.bar(searches_by_date.index, searches_by_date.values, width=0.5)
plt.xlabel('Date')
plt.ylabel('Number of Searches')
plt.title('Number of Searches by Date')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a list of columns with missing values
missing_cols = ['Part of a policing operation', 'Latitude', 'Longitude', 
                'Self-defined ethnicity', 'Legislation', 'Outcome linked to object of search', 
                'Removal of more than just outer clothing']

# Create a bar plot for each column showing the number of missing values by station
for col in missing_cols:
    missing_data = df[df[col].isna()]
    missing_data_counts = missing_data.groupby('station').size().reset_index(name='count')
    plt.figure(figsize=(12,6))
    plt.bar(missing_data_counts['station'], missing_data_counts['count'])
    plt.title(f'Number of missing values for {col} by station')
    plt.xticks(rotation=90)
    plt.show()
```

```python

# Subset the dataframe to only columns with missing values
missing_df = df[missing_cols]

# Create a dataframe that shows the proportion of missing values for each station and feature
missing_prop = pd.DataFrame(index=df['station'].unique())
for col in missing_df.columns:
    missing_prop[col] = df.groupby('station')[col].apply(lambda x: x.isna().sum() / len(x))

# Create a heatmap using seaborn
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(missing_prop, cmap='coolwarm', ax=ax)
ax.set_title('Proportion of Missing Values by Station and Feature')
plt.show()
```
